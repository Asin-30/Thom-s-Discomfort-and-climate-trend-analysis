### Feature Engineering

```python
df["temp_spread"] = df["MDBT"] - df["MDPT"]


# seasons for month
def assigne_season(month):
    if month in [12,1,2]:
        return "winter"
    elif month in [3,4,5]:
        return "summer"
    elif month in [6,7,8,9]:
        return "Monsoon"
    else:
        return "Post-Monsoon"

df['season'] = df['MN'].apply(assigne_season)

# OHE season
df= pd.get_dummies(df, columns = ['season'], drop_first = True)
# One hot encoding the season
df['is_hot_season'] = df['MN'].apply(lambda x: 1 if x in [4,5,6] else 0)
```

Predict whether a month experiences extreme heat (MDBT > 40°C) using other climate variables.

## Construction of Heat stress index : Thom's Discomfort Index

Thom's Discomfort Index: 0.5 * (MDBT + 61 + ((MDBT - 68) * 1.2) + (RH * 0.094))
where MDBT is in Farenheit and RH / MRH in %.

```python
# Converting MDBT (°C) to Fahrenheit
df1['MDBT_F'] = df1['MDBT'] * 9/5 + 32

# Computing Thom's Discomfort Index (DI) using MRH as RH (%)
df1['DI_F'] = 0.5 * (df1['MDBT_F'] + 61 + ((df1['MDBT_F'] - 68) * 1.2) + (df1['MRH'] * 0.094))

# converting Thom's DI back to Celsius 
df1['DI_C'] = (df1['DI_F'] - 32) * 5/9

# Aggregate yearly average DI for years with ≥10 months
heat_stress_trend = df1[df1['YEAR'].isin(filtered_years)].groupby('YEAR')[['DI_C']].mean().reset_index()

# Plot Thom's Heat Stress Index trend
plt.figure(figsize=(8, 4))
plt.plot(heat_stress_trend['YEAR'], heat_stress_trend['DI_C'], marker='o', color='darkred')
plt.title("Thom's Discomfort Index Trend (Annual Avg, Gwalior)")
plt.xlabel("Year")
plt.ylabel("DI (°C equivalent)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

OBSERVATION 

If the Thom's discomfort index is :
-  < 24°C can be  comfortable
-  24-26 °C can be Mild Discomfort
-  26-29 °C can be Strong Discomfort
-  exceeds 29°C can be severe heat stress.
Above plot shows that after 1983, there has been no year where DI < 29. In addition, there is a gradual rise of DI over decades.

## Heat Stress Classification

```python
def heat_stress_classification(di):
    if di < 27:
        return "Comfortable" # changeable
    elif 27 <= di < 29:
        return "Mild Discomfort"
    elif 29 <= di  < 32:
        return "Severe Discomfort"
    else: 
        return "Extreme Heat Stress"


df1['heat_stress_level'] = df1['DI_C'].apply(heat_stress_classification)
```

```python
# counting frequency of each category
print((df1['heat_stress_level'].value_counts(normalize = True)*100).sort_index())

# visualization
((df1['heat_stress_level'].value_counts(normalize = True)*100).sort_index()).plot(kind = "bar", color = "firebrick")
plt.xticks(rotation = 45)
plt.title("Monthly heat stress distribution (all years)")

plt.show()
```

This plot shows that even in non peak summer months, there are months with extreme heat stress and severe discomfort. In other words,
- There are 31% of the months which were comfortable as per DI.
- Extreme Heat stress months are ~45%. This implies that the extreme heat stress months are more frequent.
- There were 17.4% months were of severe discomfort.

This implies that Gwalior weather cause mild, severe, extreme discomfort to the population ~70% of the months.
